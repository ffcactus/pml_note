# Convolution of two Gaussians is a Gaussian

Show that the convolution of two Gaussians is a Gaussian, i.e.,
$$
p(y) = \mathcal{N}(x_1 | \mu_1, \sigma_1^2) \otimes \mathcal{N}(x_2 | \mu_2, \sigma_2^2) = \mathcal{N}(y | \mu_1 + \mu_2, \sigma_1^2 + \sigma_2^2)
$$

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
step = 0.1
x1_input = np.arange(-5, 5, step)
x2_input = np.arange(-5, 5, step)
x1_pdf = norm.pdf(x1_input, loc=-1, scale=1)
x2_pdf = norm.pdf(x2_input, loc=1, scale=1)
y = np.convolve(x1_pdf, x2_pdf, mode='same') * step
plt.plot(x1_input, x1_pdf, label='x1')
plt.plot(x2_input, x2_pdf, label='x2')
plt.plot(x1_input, y, label='y')
```
