# Deriving the inverse gamma density

Let X ~ Ga(a, b), and Y = 1/X. Derive the distribution of Y.

## Gamma distribution

The gamma distribution is a flexible distribution for positive real valued rv’s, x>0.
It is defined in terms of two parameters, called the shape a>0 and the rate b>0. Its pdf is given by
$$
Ga(x|shape=a,rate=b) \triangleq \frac{b^a}{\Gamma(a)}x^{a-1}e^{-xb}
$$
In which the $\Gamma(a)$ is defined as:
$$
\Gamma(a) \triangleq \int_0^{\infty} x^{a-1} e^{-x} dx
$$

## Change of variables: scalar case

Now let us zoom in on a point x and another point that is infinitesimally close, namely x+ dx. We see this interval gets mapped to (y,y+ dy). The probability mass in these intervals must be the same, hence p(x)dx= p(y)dy, and so p(y) = p(x)dx/dy. However, since it does not matter (in terms of probability preservation) whether
dx/dy>0 or dx/dy<0, we get
$$
p_y(y) = p_x(x) \left| \frac{dx}{dy} \right|
$$
To do the change of variables, we need do both coordination change and density change. 

Back to the question, we have y = 1/x, so 
$$
\frac{dx}{dy} = \frac{d}{dy}(y^{-1}) = - \frac{1}{y^2}
$$

So after do both coordination change and density change, and also note the the $\Gamma(a)$ is a constant, we have
$$
\begin{align}
p_y(y) &= \frac{b^a}{\Gamma(a)} (\frac{1}{y})^{a-1} e^{-\frac{b}{y}} \frac{1}{y^2} \\
       &= \frac{b^a}{\Gamma(a)} y^{-1-a} e^{-\frac{b}{y}}
\end{align}
$$
