# Variance of a sum

Show that the variance of a sum is
$$
\mathbb{V}[X + Y] = \mathbb{V}[X] + \mathbb{V}[Y] + 2\text{Cov}[X, Y]
$$
where $\text{Cov}[X,Y]$ is the covariance between X and Y.

By definition
$$
\begin{align}
\mathbb{V}[X+Y] &\triangleq \mathbb{E}[(X+Y-\mathbb{E}[X+Y])^2] \\
                &=\mathbb{E}[(X+Y-\mathbb{E}[X]-\mathbb{E}[Y])^2] \\
                &=\mathbb{E}[((X-\mathbb{E}[X])+(Y-\mathbb{E}[Y]))^2]
\end{align}
$$
Let 
$$
a=X-\mathbb{E}[X] \\
b=Y-\mathbb{E}[Y]
$$
We have
$$
\begin{align}
\mathbb{V}[X+Y] &= \mathbb{E}[a^2 + b^2 + 2ab] \\
                &= \mathbb{E}[a^2] + \mathbb{E}[b^2] + 2\mathbb{E}[ab] \\
                &= \mathbb{V}[X] + \mathbb{V}[Y] + 2\mathbb{E}[X-\mathbb{E}[X]]\mathbb{E}[Y-\mathbb{E}[Y]] \\
                &= \mathbb{V}[X] + \mathbb{V}[Y] + 2\text{Cov}[X,Y]
                
\end{align}
$$
