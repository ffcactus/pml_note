# Mean, mode, variance for the beta distribution

Suppose $\theta∼Beta(a,b)$. Show that the mean, mode and variance are given by:
$$
\mathbb{E}[\theta]=\frac{a}{a+b}
$$
$$
\mathbb{V}[\theta]=\frac{ab}{(a+b)^2(a+b+1)}
$$
$$
\text{mode}[\theta]=\frac{a-1}{a+b-2}
$$

## Beta Distribution
The beta distribution has support over the interval [0,1] and its pdf is defined as follows:
$$
\text{Beta}(x | a, b) \triangleq \frac{1}{\text{B}(a,b)} x^{a-1} (1-x)^{b-1} 
$$
where B(a,b) is the beta function, acts as the normalization, and defined by:
$$
\begin{align}
\text{B}(a,b) &\triangleq \frac{\Gamma(a) \Gamma(b)}{\Gamma(a + b)} \\
              &=\int_a^b x^{a-1} (1-x)^{b-1} dx
\end{align}
$$

## Mean or expected value
Mean is defined by:
$$
\mathbb{E}[X] \triangleq \int_{\mathcal{X}} x p(x)dx
$$
So the mean of Beta distribution is
$$
\begin{align}
\mathbb{E}[\theta] &= \frac{1}{\text{B}(a,b)} \int_0^1 x x^{a-1} (1-x)^{b-1} dx \\
                   &= \frac{1}{\text{B}(a,b)} \int_0^1 x^a (1-x)^{b-1} dx \\
                   &= \frac{1}{\text{B}(a,b)} \int_0^1 \text{B}(a+1,b) dx \\
                   &= \frac{\text{B}(a+1,b)}{\text{B}(a,b)} \int_0^1 dx \\
                   &= \frac{\text{B}(a+1, b)}{\text{B}(a,b)} \\
                   &= \frac{\frac{\Gamma(a+1)\Gamma(b)}{\Gamma(a+1+b)}}{\frac{\Gamma(a)\Gamma(b)}{\Gamma(a+b)}} \\
                   &= \frac{\Gamma(a+1)}{\Gamma(a)} \cdot \frac{\Gamma(a+b)}{\Gamma(a+b+1)}
\end{align}
$$

Using the recursive property of Gamma functions ($\Gamma(n+1) = n\Gamma(n)$):
$$
\frac{\Gamma(a+1)}{\Gamma(a)} = a \\
\frac{\Gamma(a+b)}{\Gamma(a+b+1)} = \frac{1}{a+b}
$$
So the final result is
$$
\mathbb{E}[\theta] = \frac{a}{a+b}
$$

## Variance
$$
\begin{align}
\mathbb{V}[X] &\triangleq \mathbb{E}[(X - \mu)^2] \\
              &= \mathbb{E}[X^2] - \mu^2 \\
\end{align}
$$
Let's focus on $\mathbb{E}[X^2]$:
$$
\begin{align}
\mathbb{E}[X^2] &= \int_0^1 x^2 p(x) dx \\
                &= \frac{1}{\mathbb{B}(a,b)} \int_0^1 x^{a+1} (1-x)^{b-1} dx \\
                &= \frac{\mathbb{B}(a+2,b)}{\mathbb{B}(a,b)} \\
                &= \frac{\Gamma(a+2) \Gamma(b)}{\Gamma(a+b+2)} \frac{\Gamma(a+b)}{\Gamma(a) \Gamma(b)} \\
                &= \frac{\Gamma(a+2) \Gamma(a+b)}{\Gamma(a+b+2) \Gamma(a)} \\
                &= \frac{(a+1) a}{(a+b+1)(a+b)}
\end{align}
$$
So we have
$$
\begin{align}
\mathbb{V}[X] &= \mathbb{E}[X^2] - \mu^2 \\
              &= \frac{(a+1) a}{(a+b+1)(a+b)} - \frac{a^2}{(a+b)^2} \\
              &= \frac{ab}{(a+b)^2(a+b+1)}

\end{align}
$$

## Mode

The mode of a distribution is the value with the highest probability mass or probability density:
$$
\boldsymbol{x}^* = \underset{\boldsymbol{x}}{\text{argmax}} \, p(\boldsymbol{x})
$$
The highest probability density happened when the derivation of the PDF is 0.
$$
\begin{align}
\frac{d}{dx} \text{Beta}(x | a,b) &= 0 \\
(a-1) x^{a-2} (1-x)^{b-1} - (b-1)(1-x)^{b-2} x^{a-1} &=0 \\
(a-1) (1-x) - (b-1) x &= 0 \\
\end{align}
$$
Solve the equation we have
$$
x = \frac{a-1}{a+b-2}
$$
