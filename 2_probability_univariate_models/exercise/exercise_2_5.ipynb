# Expected value of the minimum of two rv’s

Suppose X,Y are two points sampled independently and uniformly at random from the interval [0,1]. What
is the expected location of the leftmost point?

Let's introduce a new random variable $Z=min(X, Y)$. Now the question can turn to find the expectation of Z:
$$
E[Z] \triangleq \int_{0}^{1} z p(z) \, dz
$$
To find out $p(z)$, we can take the derivertive to the CDF. The CDF is
$$
Pr(0<Z<z) = 1-Pr(Z>z) = 1 - Pr(A>Z)Pr(B>Z)
$$
Oberviously, 
$$
Pr(A > z) = Pr(B > z) = 1 - z
$$
So
$$
Pr(0<Z<z) = 1-(1-z)^2 = 2z-z^2
$$
Now we need the PDF:
$$
p(z) \triangleq \frac{d}{dz}Pr(Z<z) = 2 - 2z
$$
Plugin into expectation, we have
$$
E(Z) = \int_{0}^1 (2z-2z^2) \, dz = \frac{1}{3}
$$
