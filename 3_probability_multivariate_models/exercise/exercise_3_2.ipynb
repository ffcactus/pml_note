## Correlation coefficient is between -1 and +1

The Cauchy-Schwarz Inequality states that for any two random variables (or vectors) $A$ and $B$:
$$
|\mathbb{E}[AB]| \le \sqrt{\mathbb{E}[A^2]\mathbb{E}[B^2]}
$$
Or:
$$
(\mathbb{E}[AB])^2 \le \mathbb{E}[A^2]\mathbb{E}[B^2]
$$

Now, let:
$$
A = (X - \mathbb{E}[X]) \\
B = (Y - \mathbb{E}[Y])
$$

We have:
$$
\mathbb{E}[AB] = \text{Cov}[X, Y] \\
\mathbb{E}[A^2] = \mathbb{V}[X] \\
\mathbb{E}[B^2] = \mathbb{V}[Y]
$$

So:
$$
\begin{align}
\rho &= \frac{\mathbb{E}[AB]}{\sqrt{\mathbb{E}[A^2] \mathbb{E}[B^2]}}
\end{align}
$$

Because of the Cauchy-Schwarz Inequality above, $-1 \le \rho \le 1$.
