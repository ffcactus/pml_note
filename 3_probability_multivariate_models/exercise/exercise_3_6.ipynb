# Normolization constant for a multidimentional Gaussian

Prove that the normallization constant for a d-dimentional Gaussian is given by:
$$
(2 \pi)^{d/2} {\left| \Sigma \right|}^{1/2} = \int \text{exp}(- \frac{1}{2} (x - \mu)^T \Sigma^{-1} (x - \mu)) dx
$$
Hint: diagonalize $\Sigma$ and use the fact that $\left| \Sigma \right| = \prod_i \lambda_i$ to write the joint pdf as a product of d one-dimensional
Gaussians in a transformed coordinate system. (You will need the change of variables formula.) Finally, use
the normalization constant for univariate Gaussians.


Proof: The Normalization Constant of a Multidimensional Gaussian

To prove that $(2\pi)^{d/2}|\mathbf{\Sigma}|^{1/2} = \int \exp\left(-\frac{1}{2}(\mathbf{x} - \boldsymbol{\mu})^T \mathbf{\Sigma}^{-1} (\mathbf{x} - \boldsymbol{\mu})\right)d\mathbf{x}$, we follow these four logical steps:

1. Centering the Distribution

We simplify the integral by shifting the mean to the origin. Let $\mathbf{z} = \mathbf{x} - \boldsymbol{\mu}$. Since this is a simple translation, the volume element remains unchanged ($d\mathbf{x} = d\mathbf{z}$):
$$
I = \int_{\mathbb{R}^d} \exp\left(-\frac{1}{2} \mathbf{z}^T \mathbf{\Sigma}^{-1} \mathbf{z}\right) d\mathbf{z}
$$

2. Diagonalization (Rotation)
The covariance matrix $\mathbf{\Sigma}$ is symmetric and positive-definite, so it can be eigendecomposed as $\mathbf{\Sigma} = \mathbf{U}\mathbf{\Lambda}\mathbf{U}^T$, where $\mathbf{U}$ is an orthogonal rotation matrix and $\mathbf{\Lambda}$ is a diagonal matrix of eigenvalues $\lambda_i$.

The inverse is $\mathbf{\Sigma}^{-1} = \mathbf{U}\mathbf{\Lambda}^{-1}\mathbf{U}^T$. Substituting this into our "sandwich" exponent:
$$
\mathbf{z}^T \mathbf{\Sigma}^{-1} \mathbf{z} = \mathbf{z}^T (\mathbf{U}\mathbf{\Lambda}^{-1}\mathbf{U}^T) \mathbf{z}
$$

3. Change of Variables
We define a new coordinate system $\mathbf{y} = \mathbf{U}^T\mathbf{z}$. Because $\mathbf{U}$ is a rotation matrix, the Jacobian of the transformation is $|\det(\mathbf{U})| = 1$, meaning $d\mathbf{z} = d\mathbf{y}$. The exponent simplifies to a sum of squares:
$$
\mathbf{y}^T \mathbf{\Lambda}^{-1} \mathbf{y} = \sum_{i=1}^{d} \frac{y_i^2}{\lambda_i}
$$

4. Evaluation via Fubini's Theorem

Because the variables $y_i$ are now independent (no cross-terms), the $d$-dimensional integral factorizes into a product of $d$ univariate Gaussian integrals:
$$
I = \int_{-\infty}^{\infty} \dots \int_{-\infty}^{\infty} \exp\left( \sum_{i=1}^{d} -\frac{y_i^2}{2\lambda_i} \right) dy_1 \dots dy_d
$$
$$
I = \prod_{i=1}^{d} \left( \int_{-\infty}^{\infty} e^{-\frac{y_i^2}{2\lambda_i}} dy_i \right)
$$
Using the standard 1D result $\int_{-\infty}^{\infty} e^{-ax^2} dx = \sqrt{\frac{\pi}{a}}$, where $a = \frac{1}{2\lambda_i}$:
$$
I = \prod_{i=1}^{d} \sqrt{2\pi\lambda_i} = (2\pi)^{d/2} \sqrt{\prod_{i=1}^{d} \lambda_i}
$$

**Final Result**

Since the product of the eigenvalues is the determinant of the matrix ($\prod \lambda_i = |\mathbf{\Sigma}|$):
$$
I = (2\pi)^{d/2} |\mathbf{\Sigma}|^{1/2}
$$
