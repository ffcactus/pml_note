# Correlation coefficient for linearly related variables is ±1

Show that, if $Y= aX+ b$ for some parameters $a>0$ and $b$, then $\rho(X,Y) = 1$. Similarly show that if $a<0$,
then $\rho(X,Y) =−1$.

As we know that:
$$
\mathbb{E}[X+Y] = \mathbb{E}[X] + \mathbb{E}[Y] \\
\mathbb{E}[aX+b] = a \mathbb{E}[X] + b \\
\mathbb{V}[aX+b] = a^2 \mathbb{V}[X] \\
\text{Cov}[X, Y] \triangleq \mathbb{E}[(X-\mathbb{E}[X])(Y-\mathbb{E}[Y])] = \mathbb{E}[XY] - \mathbb{E}[X] \mathbb{E}[Y]
$$

We can have:
$$
\begin{align}
\rho(X, Y) &= \frac{\text{Cov}(X,Y)}{\sqrt{a^2 {\mathbb{V}[X]}^2}} \\
           &= \frac{\mathbb{E}(a X^2 + bX) - a {\mathbb{E}[X]}^2 - b \mathbb{E}[X]}{\left| a \right| \mathbb{V}[X]} \\
           &= \frac{a(\mathbb{E}[X^2] - {\mathbb{E}[X]}^2)}{\left| a \right| \mathbb{V}[X]} \\
           &= \frac{a \mathbb{V}[X]}{\left| a \right| \mathbb{V}[X]} \\
           &= \pm 1
\end{align}
$$
