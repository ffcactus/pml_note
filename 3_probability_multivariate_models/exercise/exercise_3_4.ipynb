# Linear combinations of random variables

Let $x$ be a random vector with mean $m$ covariance matrix $\Sigma$. Let $A$ and $B$ be matrices.

a. Derive the covariance matrix of $Ax$.

b. Show that $tr(AB) = tr(BA)$.

c. Derive an expression for $\mathbb{E}[x^TAx]$.

## Solution a

As the covariance of random vector is defined as:
$$
\text{Cov}[\mathbb{\boldsymbol{x}}] \triangleq \mathbb{E}[(\boldsymbol{x} - \mathbb{E}[\boldsymbol{x}])(\boldsymbol{x} - \mathbb{E}[\boldsymbol{x}])^T] \triangleq \Sigma
$$
We have:
$$
\begin{align}
\text{Cov}[A \boldsymbol{x}] &= \mathbb{E}[(A \boldsymbol{x} - A \mathbb{E}[\boldsymbol{x}])(A \boldsymbol{x} - A \mathbb{E}[\boldsymbol{x}])^T] \\
                             &= \mathbb{E}[A(\boldsymbol{x} - \mathbb{E}[\boldsymbol{x}]) (\boldsymbol{x} - \mathbb{E}[\boldsymbol{x}])^T A^T] \\
                             &= A \mathbb{E}[(\boldsymbol{x} - \mathbb{E}[\boldsymbol{x}])(\boldsymbol{x} - \mathbb{E}[\boldsymbol{x}])^T] A^T \\
                             &= A \text{Cov}[\boldsymbol{x}] A^T
\end{align}
$$

## Solutiion b
The trace of a square matrix is the sum of the diagonal elements:
$$
tr(A) = \sum_{i=1}^{n} a_{ii} = a_{11} + a_{22} + \dots + a_{nn}
$$

Suppose $AB = C$ and $BA = D$. The elements on the diagonal of C can be expressed as:
$$
c_{ii} = \sum_{j=1}^{n} A_{ij} B_{ji}
$$
So we have:
$$
tr(C) = \sum_{i=1}^{n} c_{ii} = \sum_{i=1}^{n} \sum_{j=1}^{n} A_{ij} B_{ji}
$$
For the same reason
$$
tr(D) = \sum_{i=1}^{n} d_{ii} = \sum_{i=1}^{n} \sum_{j=1}^{n} B_{ij} A_{ji}
$$
Every single multiplication $A_{\text{row, col}} \times B_{\text{col, row}}$ exists in both sums. So they are the same.

## Solution c
$$
\begin{aligned}
\mathbb{E}[x^T Ax] &= \mathbb{E}[tr(x^T Ax)] && \text{(Since } x^T Ax \text{ is a scalar)} \\
&= \mathbb{E}[tr(Axx^T)] && \text{(By the cyclic property of the trace)} \\
&= tr(A \mathbb{E}[xx^T]) && \text{(By linearity of expectation)} \\
&= tr(A(\Sigma + mm^T)) && \text{(Since } \mathbb{E}[xx^T] = \Sigma + mm^T \text{)} \\
&= tr(A\Sigma) + tr(Amm^T) && \text{(By linearity of the trace)} \\
&= tr(A\Sigma) + m^T Am && \text{(By the cyclic property, as } m^T Am \text{ is a scalar)}
\end{aligned}
$$
